# 1.1 Climate Data via STAC

**STAC (SpatioTemporal Asset Catalog)** is the modern standard for organizing and searching geospatial data. Instead of downloading files, you query a "library" (Catalog) to find exactly what you need (Assets) based on location and time.

| Feature | **The "Old" Way** | **The "STAC" Way** |
| :--- | :--- | :--- |
| **Discovery** | Googling "Landsat download," navigating FTP sites. | Querying an API: *"Give me Sentinel-2 images of London with <10% clouds."* |
| **Access** | Downloading massive `.zip` or `.nc` files to your hard drive. | Streaming only the pixels you need (Lazy Loading) from the cloud. |
| **Standard** | Every provider has a different folder structure. | JSON metadata is standard across NASA, Microsoft, USGS, etc. |

**Learn more:** [stacspec.org](https://stacspec.org/)
